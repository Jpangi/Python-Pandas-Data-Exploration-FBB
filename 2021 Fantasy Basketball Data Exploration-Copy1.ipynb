Loading Pandas and Data

```python
import pandas as pd

nba = pd.read_csv('NBA fantasy basketball.csv')

print(nba.head())
```

```python
## Read Headers
print(nba.columns)
```

```python
## Read each column
print(nba[['PLAYER', 'GP', 'FG%']])
```

```python
# Read each Row
print(nba.iloc[1:4])
```

```python
## groupby 
nba.loc[nba['TEAM'] == "GSW"]
```

```python
## sorting by team and MPG then sorting by ascending or descending 1=asc 0=desc

nba.sort_values(['TEAM', 'MPG'], ascending=[1,0])
```

```python
# changing the 3 pointers made column from 3:00pm to 3PM
nba = nba.rename(columns={'3:00 PM': '3PM'})
nba
```

```python
# splitting FG% and FT% into only the percentage

new = nba['FG%'].str.split("(", n = 1, expand = True)
nba['FG_percent'] = new[0]
nba.drop(columns = ['FG%'], inplace = True)

new = nba['FT%'].str.split("(", n = 1, expand = True)
nba['FT_percent'] = new[0]
nba.drop(columns = ['FT%'], inplace = True)

nba
```

```python
# lets see if we can find players in the top 75% for multiple categories
display(nba.loc[(nba['PTS']>1230.250000) & (nba['AST']>303.250000)& (nba['FT_percent']> 0.853)& (nba['3PM']> 157.000000)&(nba['STL']>75.000000 ),
                    ['PLAYER']])
```

6 players are in the top 75% in 5 categories(PTS, AST, FT%, 3PM and STL). It would be incredible if we could make a team out of these 6 players but more then likely if you're playing fantasy basketball you'd never be able to get all of these players on your roster. However it's a great way to see what players match well together on a team.

# Exploring player Positions

```python
#Quick view of the PG/SG position
POS_PG =nba.query("POS == 'PG' or POS =='PG,SG'") 
print(POS_PG)
```

```python
nba['POS'].value_counts()
```

From this we can see that PG,SG position is the most frequent position in the top 200 players. Even more suprising is that there is only one true SG in the top 200 players. If you're looking for a good value pick you could grab one of the 17 players that can play 3 positions on the court.
